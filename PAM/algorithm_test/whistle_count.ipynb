```python
import pandas as pd
w = pd.read_csv("whistle_features_manwork_right.csv")
```

```python
time_slice = 2*60
length = 0
count = 0
whistle_count = []
for index, row in w.iterrows():
    start_time = int(row["Start Time"])
    if(start_time//time_slice==length):
        count+=1
    else:
        whistle_count.append(count)
        count = 1
        length += 1
whistle_count.append(count)
```
