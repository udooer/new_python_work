```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
fs = 96000
#
fft_number = 1024
overlab = 0.5
window_type = 'hann'
#
hop_size = math.ceil(fft_number*(1-overlab))
half_size = math.ceil((fft_number+1)/2)
#
start_fre = 3000
end_fre = 10000

threshold = 2
frequency_width = 300

df = fs/fft_number

f = np.arange(half_size)*fs/fft_number
df = fs/fft_number
start_index = math.floor((start_fre)/df)
end_index = math.ceil((end_fre)/df)
width_size = math.ceil(frequency_width/2/df)
```

```python
with open("detection.npy","rb") as f:
    detection = np.load(f)
detection.shape
```

```python
fig, ax = plt.subplots(figsize=(50,5))
im = ax.imshow(detection, cmap='binary',  origin='lower', aspect='auto',
               interpolation='hanning', extent=(0, 5, f[start_index]/1000, f[end_index]/1000))
ax.set_ylim([0,20])
ax.set_xlabel("Time(s)")
ax.set_ylabel("Frequency(KHz)")
c_bar = fig.colorbar(im)
plt.show()
```
